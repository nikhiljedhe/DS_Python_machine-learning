```python
# Convert the dataframe into a list of lists for TransactionEncoder
# Each row in the dataframe is a transaction, and items are comma-separated
transactions_list = df.iloc[:,0].str.split(',').tolist()
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

te = TransactionEncoder()
te_array = te.fit(transactions_list).transform(transactions_list)
dataset = pd.DataFrame(te_array, columns=te.columns_)
```

```python
transactions_list = df.iloc[:, 0].str.split(',').tolist()
te = TransactionEncoder()
te_array = te.fit(transactions_list).transform(transactions_list)
dataset = pd.DataFrame(te_array, columns=te.columns_)
```

```python
frequent_itemsets = apriori(dataset, min_support=0.01, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
display(rules)
```

```python
# Sort rules by lift in descending order to see the strongest positive associations first
high_lift_rules = rules.sort_values(by='lift', ascending=False)
display(high_lift_rules.head(10))

# Filter rules with a higher confidence threshold to identify more reliable associations
high_confidence_rules = rules[rules['confidence'] > 0.5].sort_values(by='confidence', ascending=False)
display(high_confidence_rules)
```

1.  **Support, Confidence, and Lift in Association Rule Mining:**
    -   **Support:** Measures how frequently an itemset appears in the dataset. It's the proportion of transactions containing the itemset.
    -   **Confidence:** Indicates the reliability of a rule X -> Y, showing how likely Y is purchased when X is purchased. It's the proportion of transactions containing X that also contain Y.
    -   **Lift:** Measures how much more likely Y is purchased when X is purchased compared to the probability of purchasing Y independently. A lift > 1 indicates a positive association, < 1 a negative association, and = 1 no association.

2.  **Limitations of Association Rule Mining:**
    *   High computational cost for large datasets.
    *   Generation of numerous trivial or non-actionable (spurious) rules.
    *   Difficulty handling continuous data (requires discretization).
    *   Challenges in finding associations for rare items due to minimum support thresholds.
    *   Interpretability issues with rules involving many items.
    *   Standard algorithms often ignore item quantities and purchase order.

### Data Analysis Key Findings

*   Applying the Apriori algorithm with a minimum support of 0.01 identified frequent itemsets.
*   Association rules were generated from the frequent itemsets, filtered to include rules with a minimum confidence of 0.5.

*   An attempt to filter rules with higher confidence (> 0.6) and support (> 0.02) resulted in no rules meeting these stricter criteria simultaneously in this dataset.
