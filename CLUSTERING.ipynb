```python
X = df[[ 'Balance', 'Qual_miles', 'cc1_miles', 'cc2_miles','cc3_miles', 'Bonus_miles', 'Bonus_trans', 'Bonus_trans.1','Flight_trans_12', 'Days_since_enroll', 'Award?']]
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X)
X_scaled=scaler.transform(X)
X_scaled
```

#K-means Clustering

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=6)
kmeans.fit(X_scaled)
kmeans_labels=kmeans.labels_
kmeans_labels
```

```python
pred=kmeans.predict(X_scaled)
pred
```

```python
wcss =[]
for i in range(1,11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("ELBOW CURVE")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
```

```python
# Visualizing the clusters (using the first two features for simplicity)
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans_labels, cmap='viridis', s=50)
plt.title('K-Means Clustering (First Two Features)')
plt.xlabel('Scaled Balance')
plt.ylabel('Scaled Qual_miles')
plt.colorbar(label='Cluster Label')
plt.show()
```

```python
df['Cluster'] = pred
df.head()
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(X_scaled, pred)
print(f"Silhouette Score kmeans clustering : {silhouette_avg}")
```

#Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
h_cluster = AgglomerativeClustering(n_clusters = 6)
h_cluster.fit(X_scaled)
```

```python
pred1=h_cluster.fit_predict(X_scaled)
pred1
```

```python


from scipy.cluster.hierarchy import linkage, dendrogram

# Generate the linkage matrix
linked = linkage(X_scaled, 'ward')

# Plot the dendrogram
plt.figure(figsize=(15, 7))
dendrogram(linked,
            orientation='top',
            distance_sort='descending',
            show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data points')
plt.ylabel('Distance')
plt.show()
```

```python
silhouette_avg1 = silhouette_score(X_scaled, pred1)
print(f"Silhouette Score hierarchical clustering: {silhouette_avg1}")
```

#DBSCAN

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=1,min_samples=8)
dbscan.fit(X_scaled)
```

```python
pred2=dbscan.fit_predict(X_scaled)
pred2
```

```python
silhouette_avg2 = silhouette_score(X_scaled, pred2)
print(f"Silhouette Score DBSCAN clustering: {silhouette_avg2}")
```

#Silhouette Score Interpretation
The silhouette score is a measure of how similar an object is to its own cluster (cohesion) compared to other clusters (separation). The score ranges from -1 to 1.

- K-Means Silhouette Score: 0.243: This score indicates that the data points in the K-Means clusters are somewhat similar to their assigned cluster, but there might be some overlap or noise in the data. A score of 0.243 suggests a weak structure.
- DBSCAN Silhouette Score: 0.211: This score is slightly lower than the K-Means score, indicating that the DBSCAN clusters might have more noise or overlap. However, DBSCAN is designed to handle noise and outliers, so this score is still reasonable.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.cluster import KMeans

param_grid = {'n_clusters': range(2, 11)}
grid_search = GridSearchCV(KMeans(), param_grid, cv=5, scoring=lambda estimator, X: estimator.score(X))
grid_search.fit(X_scaled)

print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
```
