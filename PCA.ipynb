```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generate boxplots for each column (excluding the first 'Type' column) grouped by the 'Type' column
# This helps visualize the distribution of features for each wine type
for c in df.columns[1:]:
    df.boxplot(c,by='Type',figsize=(7,4),fontsize=14)
    plt.title("{}\n".format(c),fontsize=16)
    plt.xlabel("Wine Type", fontsize=16)
```

```python
# Generate histograms for each column to visualize the distribution of values
for col in df.columns:
  plt.hist(df[col], bins=50)
  plt.title(f"histogram of {col}")
  plt.show()
```

```python
# Calculate the correlation matrix of the DataFrame, considering only numeric columns
import seaborn as sns
import matplotlib.pyplot as plt
df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
# Initialize a StandardScaler to standardize the features (mean=0, variance=1)
# Apply the StandardScaler to the features (excluding the 'Type' column)
# Store the scaled data in a new DataFrame with original column names
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_scaled=pd.DataFrame(scaler.fit_transform(df.iloc[:,1:]),columns=df.columns[1:])
df_scaled.head()
```

```python
df_scaled.describe().transpose()
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=13)
df_pca=pca.fit(df_scaled)
```

```python
eigenvalues = pca.explained_variance_ratio_
plt.plot(eigenvalues)
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance')
plt.show()
```

```python
# Select the optimal number of principal components
n_components = 2  # Based on the plot
pca = PCA(n_components=n_components)
pca_components = pca.fit_transform(df_scaled)


print(f"Original shape: {df_scaled.shape}")
print(f"Transformed shape: {df_pca.shape}")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Apply K-means clustering
kmeans = KMeans(n_clusters=3)
labels = kmeans.fit_predict(df)

# Reduce dimensionality using PCA for visualization
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df)

# Visualize the clustering results
plt.figure(figsize=(8, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=labels, cmap='viridis')
plt.title('K-means Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Apply K-means clustering to df_scaled
kmeans_scaled = KMeans(n_clusters=3)
labels_scaled = kmeans_scaled.fit_predict(df_scaled)

# Evaluate the clustering performance
silhouette_scaled = silhouette_score(df_scaled, labels_scaled)
davies_bouldin_scaled = davies_bouldin_score(df_scaled, labels_scaled)
print(f"Silhouette Score (df_scaled): {silhouette_scaled:.3f}")
print(f"Davies-Bouldin Index (df_scaled): {davies_bouldin_scaled:.3f}")
```

Silhouette Score
- Silhouette Score (df_scaled): 0.285
- This score indicates that the clusters are not well-separated and may have a significant overlap.
- A score of 0.285 suggests that the clustering structure is weak.

Davies-Bouldin Index
- Davies-Bouldin Index (df_scaled): 1.389
- This index indicates that the clusters have a relatively high ratio of within-cluster distances to between-cluster distances.
- A score of 1.389 suggests that the clusters are not compact or well-separated.

```python

# Apply K-means clustering to df_pca
kmeans_pca = KMeans(n_clusters=3)
labels_pca = kmeans_pca.fit_predict(df_pca)

# Evaluate the clustering performance
silhouette_pca = silhouette_score(df_pca, labels_pca)
davies_bouldin_pca = davies_bouldin_score(df_pca, labels_pca)
print(f"Silhouette Score (df_pca): {silhouette_pca:.3f}")
print(f"Davies-Bouldin Index (df_pca): {davies_bouldin_pca:.3f}")
```

Silhouette Score
- Silhouette Score (df_pca): 0.572
- This score indicates that the clusters are reasonably well-separated and have a moderate level of cohesion.
- A score of 0.572 suggests that the clustering structure is fairly strong.

Davies-Bouldin Index
- Davies-Bouldin Index (df_pca): 0.533
- This index indicates that the clusters are relatively compact and well-separated.
- A score of 0.533 suggests that the clusters have a good balance between within-cluster similarity and between-cluster dissimilarity.


Silhouette Score
- df_scaled: 0.285 (weak clustering structure)
- df_pca: 0.572 (reasonable clustering structure)

The silhouette score indicates that the clustering structure is much stronger in the df_pca dataset.

Davies-Bouldin Index
- df_scaled: 1.389 (clusters not compact or well-separated)
- df_pca: 0.533 (clusters relatively compact and well-separated)

The Davies-Bouldin index also suggests that the clusters are more compact and well-separated in the df_pca dataset.

Implications
The clustering performance metrics indicate that applying PCA improves the clustering results. The df_pca dataset has a stronger clustering structure and more compact clusters.

- Clustering Algorithm: K-means clustering can be effective in identifying clusters in the PCA-transformed data, but the choice of algorithm and parameters may depend on the specific problem and dataset.
