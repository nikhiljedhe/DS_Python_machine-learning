1. What is the difference between precision and recall?
Precision and recall are two important metrics used to evaluate the performance of a classification model.

- Precision: Precision measures the proportion of true positives among all predicted positive instances. It answers the question: "Out of all instances predicted as positive, how many are actually positive?"
- Recall: Recall measures the proportion of true positives among all actual positive instances. It answers the question: "Out of all actual positive instances, how many were predicted as positive?"

In other words:

- Precision focuses on the accuracy of positive predictions.
- Recall focuses on the ability of the model to detect all positive instances.

The formulas for precision and recall are:

- Precision = TP / (TP + FP)
- Recall = TP / (TP + FN)

Where:

- TP: True Positives
- FP: False Positives
- FN: False Negatives

2. What is cross-validation, and why is it important in binary classification?
Cross-validation is a technique used to evaluate the performance of a machine learning model by training and testing it on multiple subsets of the available data.

In binary classification, cross-validation is important because it:

- Reduces overfitting: By training and testing the model on different subsets of data, cross-validation helps to prevent overfitting and provides a more accurate estimate of the model's performance.
- Provides a robust evaluation: Cross-validation gives a more robust evaluation of the model's performance by reducing the impact of randomness in the data split.
- Helps in hyperparameter tuning: Cross-validation can be used to tune hyperparameters by evaluating the model's performance on different subsets of data.

Common types of cross-validation include:

- K-fold cross-validation: The data is split into k subsets, and the model is trained and tested k times, each time on a different subset.
- Stratified cross-validation: The data is split into subsets while maintaining the same proportion of positive and negative instances in each subset.

By using cross-validation, you can get a more accurate estimate of your model's performance and improve its reliability./

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import pickle


with open('model.pkl', 'rb') as f:
    model = pickle.load(f)

st.title("Titanic Survival Prediction")
st.subheader("Survival Prediction using Logistic Regressor")
st.header('Input Features')
Pclass = st.selectbox('Pclass', [1, 2, 3])
Sex = st.selectbox('Sex', ['male', 'female'])
Age = st.slider('Age', 0, 80, 30)
SibSp = st.slider('SibSp', 0, 8, 0)
Parch = st.slider('Parch', 0, 6, 0)
Fare = st.number_input('Fare', 0, 100)
Cabin = st.number_input('Cabin', 0, 150)
Embarked = st.selectbox('Embarked', ['S', 'C', 'Q'])

sex_encoded = 1 if Sex == 'male' else 0
embarked_encoded = {'S': 0, 'C': 1, 'Q': 2}[Embarked]

input_data = pd.DataFrame({
    'Pclass': [Pclass],
    'Sex': [sex_encoded],
    'Age': [Age],
    'SibSp': [SibSp],
    'Parch': [Parch],
    'Fare': [Fare],
    'Cabin': [Cabin],
    'Embarked': [embarked_encoded]
})

prediction = model.predict(input_data)
survival = 'Survived' if prediction[0] == 1 else 'Did not Survive'

st.write(f'Prediction: {survival}')
```
